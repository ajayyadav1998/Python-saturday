```python
file = open('table.txt')
data = file.read()
file.close()
```

```python
data
```

```python
data = data.split('\n')
```

```python
data
```

```python
toda = list(map(lambda string : string.split(),data))
print(toda)
```

```python
from functools import reduce
```

```python
b = []
for var in toda:
    b.extend(var)
```

```python
b
```

```python
c = []
for var in b:
    c.append(int(var))
    
print(c)
```

```python
re = reduce(lambda x,y : x+y ,c)
print(re)
```

```python
file = open('target.txt')
data = file.read()
file.close()
```

```python
data
```

```python
data = data.split('\n')
```

```python
data
```

```python
t = list(map(lambda string : string.split(),data))
t
```

```python
ma = []
py = []
ch = []
for p,q,r,s in t:
    ma.append(q[:])
    py.append(r[:])
    ch.append(s[:])
```

```python
ma.remove('math')
py.remove('phys')
ch.remove('chem')
```

```python
print(ma,py,ch,sep='\n')
```

```python
m = ma + py + ch
print(m)
```

```python
t = []
for var in m:
    t.append(int(var))
    
t
```

```python
r = reduce(lambda x,y : x+y,t)
r
```
