# Top 10 companies bar plot

```python
plt.figure(figsize=(10,5),facecolor='indianred')
df.companies.value_counts()[:10].plot(kind='bar',title='Top 10 companies')
plt.xlabel('Companies')
plt.ylabel('Number of Jobs')
plt.show()
```

# case status pie plot

```python
cs = df.case_status.value_counts()[:-3]
```

```python
plt.figure(figsize=(10,5))
plt.pie(x=cs.values,labels=cs.index,autopct='%.f%%',explode=[0.2,0,0,0],shadow=True)
plt.title('Case-status')
plt.show()
```

# pie plot of full time position

```python
plt.figure(figsize=(10,5))
df.ftposition.value_counts().plot(kind='pie',autopct='%.f%%',colors=['r','g'])
plt.title('Full Time Position Pie Plot')
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='indianred')
df.nlargest(10,'wage')[['companies','wage']].set_index('companies').plot(kind='barh')
plt.show()
```

```python
df['worksite'] = df.worksite.apply(lambda x : x.split(',')[1])
```

```python
df.head()
```

```python
top10work = df.worksite.value_counts()[:10]
```

```python
top10denied = df[df.case_status == 'DENIED']['worksite'].value_counts()[:10]
```

```python
top10work
```

```python
color = ['r','g','b','pink','m','k','yellow','orange','purple','violet']
```

```python
fig = plt.figure(figsize=(15,8),facecolor='yellow')

ax1 = fig.add_axes([0,0,0.45,0.5]) #left,bottom,width,height
ax2 = fig.add_axes([0.5,0,0.45,0.5])


ax1.bar(x=top10work.index,height=top10work[:],color=color)
ax1.set_title('top 10 worksite')
ax1.set_xticklabels(labels=top10work.index,rotation=45)
ax1.set_ylabel('number of values')

ax2.bar(x=top10denied.index,height=top10denied,color=color)
ax2.set_title('top 10 denies worksite')
ax2.set_xticklabels(labels=top10denied.index,rotation=45)
ax2.set_ylabel('number of values')

plt.show()
```
