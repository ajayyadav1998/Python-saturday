```python
ndf = pd.DataFrame([[randint(1,100) for var in range(50)],
                   [randint(1,200) for var in range(50)],
                   [randint(1,200) for var in range(50)]])
```

```python
ndf = ndf.T  #transpose
```

```python
ndf.columns = ['one','two','three']
```

```python
ndf.head()
```

```python
ndf.tail()
```

```python
ndf.sort_values('one',inplace=True)
```

```python
ndf.reset_index(inplace=True)
```

```python
ndf.head()
```
